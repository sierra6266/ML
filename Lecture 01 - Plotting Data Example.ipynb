# Lecture 1 - Data Plotting Examples

**Objective:** 
To learn how to plot simple scatter plots of some data

## Example 1

- Set a random seed to 17
- Generate vector $x_1$ containing 1000 randomly generated numbers
- Generate vector $y_1$ as a vector of 1000 randomly generated numbers
- plot x_1 against $y_1$

```r
set.seed(11)
x1 <- rnorm(1000)
y1 <- x1 + rnorm(1000)
plot(x1, y1)
```

## Example 1

- use the plot() function to plot $x_1$ against $y_1$
- draw a regression line (in red) fitted to those points

```r
plot(x1, y1)                                   # Apply plot function
abline(lm(y1 ~ x1), col = "red")               # Draw regression line
```

## Example 2
- Generate 1000 points of a binomial distribution of probability of 0.1 (group 1)
- Create a plot of $x_1$ and $y_1$ with their labels as the two groups above

```r
# Generate the groups first

group <- rbinom(1000, 1, 0.7)+1
cat(group)
```

```r
# check the sum of the group to ensure probability is 30%

cat("The ratio of 1s:",sum(group==1)/1000, "the ratio of 2s",sum(group==2)/1000)
```

```r
plot(x1, y1,                                   # Create plot with groups
     main = "This is my Plot",
     xlab = "X-Values",
     ylab = "Y-Values",
     col = group,
     pch = group,
     cex=2)
```
